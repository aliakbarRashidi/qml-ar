```python
from skimage.io import imread
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from skimage.draw import circle
```

```python
I = imread('./images/image.jpg')
```

```python
H = np.array([float(x) for x in """
0.363791 -0.754746   496.795         0         
   0.80062  0.170392   9.52074         0         
0.000445797 1.48909e-05         1         0         
         0         0         0         0         
""".split()]).reshape(4, 4)[:3,:3]
```

```python
image_point = H @ np.array([600, 600, 1]).reshape(3, 1)
image_point = image_point / image_point[2]
```

```python
I1 = np.copy(I)
I1[circle(359-image_point[1, 0], image_point[0, 0], 10)] = 255
```
